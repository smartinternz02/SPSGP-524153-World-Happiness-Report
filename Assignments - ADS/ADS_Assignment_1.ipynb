```python
from datetime import date, timedelta

start_dt = date(2023, 2, 1)
end_dt = date(2023, 3, 1)


delta = timedelta(days=1)


dates = []

while start_dt <= end_dt:
    
    dates.append(start_dt.isoformat())
    
    start_dt += delta

print('Dates between', start_dt, 'and', end_dt)
print(dates)
```
